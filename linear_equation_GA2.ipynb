```python
import numpy as np
import random
random.seed(100)
```

```python
fit_chromosomes=[]
population=[[random.randint(1,10) for i in range(6)] for j in range(6)]
print(population)
```

```python
def fitness(population,fit_chromosomes):
  fitness_score=[]
  for chromosome in population:
        individual_interval=abs(30-(chromosome[0]+2*chromosome[1]-3*chromosome[2]+chromosome[3]+4*chromosome[4]+chromosome[5]))
        individual_fitness=1/(1+individual_interval)
        fitness_score.append(individual_fitness)
  for i in range(len(fitness_score)):
        if fitness_score[i]==1:
            if population[i] not in fit_chromosomes:
                fit_chromosomes.append(population[i])
  return fitness_score

fitness_score=fitness(population,fit_chromosomes)
print('Fit chromosomes:',fit_chromosomes,'\n')
print('Population:',population,'\n')
print('Fitness score',fitness_score)
```

```python
def selection(population, fit_chromosomes,fitness_score):

    total_fitness = sum(fitness_score)

    p=[]
    for f in fitness_score:
        fitness_probability=f/total_fitness
        p.append(fitness_probability)

    cdf = np.cumsum(p)

    R=np.random.uniform(0,1,len(population))

    selected_chromosomes=[]
    for r in R:
        for i in range(len(population)-1):
            if r<=cdf[i]:
                selected_chromosomes.append(population[0])
                break
            if cdf[i]<r and r<=cdf[i+1]:
                selected_chromosomes.append(population[i+1])
                break

    print('p',p)
    print('cdf',cdf)
    print('R',R,'\n')

    return selected_chromosomes

print('Population before selection:',population,'\n','Size:',len(population),'\n')
selected_chromosomes=selection(population, fit_chromosomes, fitness_score)
print('Population after selection:',selected_chromosomes,'\n','Size:',len(selected_chromosomes),'\n')
```

```python
def crossover(selected_chromosomes):
    crossover_rate=0.85

    parents=[]
    for i in range(len(selected_chromosomes)):
        r=np.random.uniform(0,1,1)
        if r<crossover_rate:
          parents.append(selected_chromosomes[i])


    crossover_children = []
    for k in range(len(parents)):
        crossover_point=random.randint(1,3)
        p1_idx = k%len(parents)
        p2_idx = (k+1)%len(parents)
        crossover_children.append(parents[p1_idx][0:crossover_point]+parents[p2_idx][crossover_point:])

    print('parents',parents)
    print('crossover children',crossover_children,'\n')

    return crossover_children

print('Population before crossover:',selected_chromosomes,'\n','Size:',len(selected_chromosomes),'\n')
crossover_children=crossover(selected_chromosomes)
#population=selected_chromosomes+crossover_children
print('Population after crossover:', population,'\n','Size:',len(population),'\n')
```

```python
def mutation(crossover_children):
    mutation_rate=0.1

    genes=[]
    for chromosome in crossover_children:
      for gene in chromosome:
        genes.append(gene)

    mutation_children=[]
    for i in range(len(genes)):
        r=np.random.uniform(0,1)
        if r<mutation_rate:
          row=int(((i+1)/6)-1)
          column=int(i%6)
          crossover_children[row][column]=int(np.random.randint(0,10,1))
          mutation_children.append(crossover_children[row])

    print('mutation children',mutation_children,'\n')

    return mutation_children

print('Population before crossover:',population,'\n','Size:',len(population),'\n')
mutation_children=mutation(crossover_children)
#population=population+mutation_children
print('Population after crossover:', population,'\n','Size:',len(population),'\n')
```

```python
def get_fit_chromosomes(generations,population,fit_chromosomes):

    for generation in range(generations):
        generation+=1
        fitness_score=fitness(population,fit_chromosomes)
        selected_chromosomes=selection(population,fit_chromosomes,fitness_score)
        crossover_children=crossover(selected_chromosomes)
        population=selected_chromosomes+crossover_children
        mutation_children=mutation(crossover_children)
        population=population+mutation_children

    return fit_chromosomes

solutions=get_fit_chromosomes(10,population,fit_chromosomes)
print('Solutions:', solutions,'\n','Size:',len(solutions),'\n')
```
